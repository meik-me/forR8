# Project Title: "Predict chemical compounds close to the fibre surface in the hardened cement paste using ML classification methods"

# 1. Introduction

The micro-structure of the hardened cement paste constituting the concrete is heterogeneous, especially in the vicinity of the inclusions, such as aggregates and/or fibres. The transition zone developing between the inclusion and the cement paste is usually more porous compared to the one in the bulk material. The transition zone is important for the bond strength and the bearing capacity of the final composite. 

One of the main principal hydration products of cement is calcium silicate hydrate (CSH) gel and calcium hydroxide (CH). CSH gel forms the bond between the inclusion and the cement past. Large platy or prismatic CH crystals can enrich on the inclusion surface, and they usually crystallise within the water-filled large capillaries and/or interfacial pores. The size of well formed CH crystals is two or three orders of magnitude larger than the CSH gel particles. In contrast to CSH gel, the massive crystals of CH have low binding properties.

Providing the high bond strength between the inclusion and the cement paste would need dense CSH gel close to the inclusion surface, rather than the massive and brittle CH crystals.

# 2. Problem Formulation

The compounds in the cement paste can be estimated based on the ratio of the chemical elements measured by the SEM-EDX analysis. SEM-EDX analysis are using the X-rays to scan the material surface, e.g. a point, and estimate the chemical elements present within this discrete infinitely small area--point. For example, to define the presence of the CSH gel in the cement paste, the ratio between the Si/Ca is calculated.


In the present study, the machine learning (ML) problem was formulated as follows: \\
Predict the chemical compounds close to the fibre surface using the discrete scans of the SEM-EDX method, as the data points. The distance from the fibre surface and the ratios of the chemical elements, as the features. The labels (quantity of interest) are the reference ratios of the Si/Ca in the CH and CSH gel received by the EDX individual micro-analysis of these compounds and reported in the previous similar studies \cite{eik.rsf.jact.2019}.

SEM-EDX line analysis were made along the lines starting at the fibre surface and ending at the distance of 60 um from the fibre. Essentially, the line analysis consist of the row of discrete points with a step of 0.5 um, meaning that each discrete point is an infinitesimal area from where the chemical elements are measured. ## Labels: reference ratios of the Si/Ca in the CH and CSH gel

\begin{itemize}
\item CH: when the Si/Ca ratio is less than 0.14, i.e. 0.034-0.14;
\item CSH gel: when the Si/Ca ratio is between 0.25 and 0.53, i.e  0.25 < Si/Ca<0.53.
\end{itemize}

<img src="labels_multi.png" width="600" height="400" alt="">

```python
def multi_labels():
    """ 
    :return: array-like, shape=(m, 1), label-vector
    """ 
    cl = pd.read_csv("./data/train_val/train_val.csv",sep = '\t')
    cl=cl.rename(columns={"Si_Ca": "CSH"})
    cl=cl.CSH.values.reshape(-1,1)
    m = cl.shape[0]       # set m equal to the number of rows in features  
    y = np.zeros((m, 1));    # initialize label vector with zero entries  

    for i in range(m):
        if (cl[i] > 0.25 and cl[i] < 0.53):
            y[i,:] = 0  # Class 0 -> 0.25<Si/Ca<0.53
        elif (cl[i] > 0.14 and cl[i] <0.24):
            y[i,:] = 1  # Class 1 -> 0.14<Si/Ca<0.25
        elif (cl[i] > 0.034 and cl[i] <0.14):
            y[i,:] = 2  # Class 2 -> 0.034<Si/Ca<0.14    
        else:
            y[i,:] = 3 #Class 3 --> Si/Ca>0.53
    return y
```

```python
def multi_labels_test():
    """ 
    :return: array-like, shape=(m, 1), label-vector
    """ 
    cl = pd.read_csv("./data/test/test.csv",sep = '\t')
    cl=cl.rename(columns={"Si_Ca": "CSH"})
    cl=cl.CSH.values.reshape(-1,1)
    m = cl.shape[0]       # set m equal to the number of rows in features  
    y = np.zeros((m, 1));    # initialize label vector with zero entries  

    for i in range(m):
        if (cl[i] > 0.25 and cl[i] <0.53):
            y[i,:] = 0  # Class 0 -> 0.25<Si/Ca<0.53
        elif (cl[i] > 0.14 and cl[i] <0.24):
            y[i,:] = 1  # Class 1 -> 0.14<Si/Ca<0.25
        elif (cl[i] > 0.034 and cl[i] <0.14):
            y[i,:] = 2  # Class 2 -> 0.034<Si/Ca<0.14    
        else:
            y[i,:] = 3 #Class 3 --> Si/Ca>0.53
    return y
```

```python
def feature_matrix():
    """
    return: array-like, shape=(m, n), feature-matrix with n features m data points 
    """
    feat = pd.read_csv("./data/train_val/train_val.csv",sep = '\t')
    feat=feat.rename(columns={"Si_Ca": "CSH", "Al_Ca": "AF_tm"})
    feat=feat[["dist_um","CSH","AF_tm"]].values
    n=feat.shape[1] # set n to the number of colums 
    m=feat.shape[0] # set m equal to the number of rows
    
    
    X = np.zeros((m,n))
    for i in range(m):
        X[i] = np.array(feat[i])
    return X
```

# 3. Methods

It was assumed that the relationship between the features and the labels was not linear. The features were linked to the lables depending on the Si/Ca ratio, which defined the respective class. Altogether there were 4 classes.

The data, i.e. 3035 data points, were split to training and test sets. The training data included 2601 data points, whereas the test set was 434 data points. 
While searching for an appropriate hypothesis, the following models were tested:
\begin{itemize}
\item Multi-class Logistic Regression. It was employed such, as the one classifier was trained for each class, i.e. one versus all.
\item Regularisation with Ridge and Lasso regressions.
\item Decision Tree classifier was also used to examine the perfomance on the data set.
\item Artificial Neural Network was also applied to the data.
\end{itemize}

The quality of each hypotheis space mentioned was evaluated using the (average) squared error loss???. The accuracy of the resulting classifier was calculated as the fraction of correctly predicted classes, i.e. for which $\mathrm{y^{(i)} =\hat{y^{(i)}}}$. The used accuracy is an appropriate measure for the model performance, since the label numbers were mapped based on the class assigned.

The best ML model was selected employing the K-fold cross validation, meaning that the training data (2601 data points) was split into the training and validation sets, and the fold with the minimum validation error was chosen. ## 3.2 K-fold Cross-Validation: splitting the data into training/validation sets

The main feature of the K-Fold approach is that each data point appears exactly once in the test (validation) set, with no repetition.
### For Multiclass Classification

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.model_selection import train_test_split
```

```python
#df_train=np.array(df_train_scal)
df_train=np.array(df_train_norm)

y_mul=multi_labels()
y_mul=y_mul.ravel()

print('Data for K-training/validation splits:', df_train.shape)
print('Labels for K-training/validation splits:',y_mul.shape)
```

```python
from sklearn.model_selection import KFold

K=4    # Specify the number of folds of split data into
kf = KFold(n_splits=K, shuffle=True)    
#The "shuffle" is mainly useful if your data is somehow sorted by classes, because then each fold might 
#contain only samples from one class (in particular for stochastic gradient decent classifiers sorted classes
#are dangerous). For other classifiers, it should make no differences. If shuffling is very unstable, your 
#parameter selection is likely to be uninformative (aka garbage).
#“shuffle” keywords that indicates whether we want a random permutation of the entries before splitting, 
#and the (optional) "random_state=1" for reproducibility.

iteration = 0
for train_indices, test_indices in kf.split(df_train):#df_train should be np.array
    iteration += 1
    df_tra_mult = df_train[train_indices,:]    # Indices here are the row numbers; Get the training set    
    df_val_mult = df_train[test_indices,:]    # Get the validation set
    print('Iteration {}:'.format(iteration))
    print('Indices for validation set:', test_indices)
    print('Indices for training set:', train_indices)
    print('df_val_mult shape: {}, df_tra_mult shape: {} \n'.format(df_val_mult.shape, df_tra_mult.shape)) 
```

```python
m = len(df_train[:,0])   
n = len(df_train[0]) 

K_list = list(range(2,11))   # List of K:s to use (2,...,10)
err_train_folds_mul = []  # List to store the average training error over K folds for each K in K_list
err_val_folds_mul = []  
```

```python
# Loop over all values of K in K_list
for K in K_list:

    train_errors_per_cv_iteration = [] #List to store the training errors of the 'K' train/val splits (for current K)
    val_errors_per_cv_iteration = []  
    
    kf = KFold(n_splits=K, shuffle=True)    
 
    for train_indices, val_indices in kf.split(df_train):
        logreg_mult = LogisticRegression(solver='liblinear', random_state=0, multi_class="ovr")
        #NO REGULARISATION if penalty='none',C:float, default=1.0; penalty is a string ('l2' by default)
        logreg_mult = logreg_mult.fit(df_train[train_indices,:], y_mul[train_indices])
        
        y_hat_train_mul = logreg_mult.predict(df_train[train_indices,:])#.reshape(-1, 1)
        train_errors_per_cv_iteration.append(metrics.accuracy_score(y_mul[train_indices], y_hat_train_mul))
        
        y_hat_val_mul = logreg_mult.predict(df_train[val_indices,:])
        val_errors_per_cv_iteration.append(metrics.accuracy_score(y_mul[val_indices], y_hat_val_mul))    # Add the test error to list of errors
     
    
    err_train = np.mean(train_errors_per_cv_iteration)    # compute the mean of round-wise training errors
    err_val = np.mean(val_errors_per_cv_iteration)    # compute the mean of round-wise validation errors
    err_train_folds_mul.append(err_train)
    err_val_folds_mul.append(err_val) 
```

```python
print('Training accuracy of Multi-class Logistic Reg. for each K:')
print(err_train_folds_mul, '\n')
print('Validation accuracy of Multi-class Logistic Reg. for each K:')
print(err_val_folds_mul, '\n')
```

```python
# Plot the training and validation errors as a function of the number K of folds
plt.figure(figsize=(8,6))
plt.plot(K_list, err_train_folds_mul)
plt.plot(K_list, err_val_folds_mul)
plt.xlabel('K folds')
plt.ylabel("ACCURACY")
plt.title("Average training and validation errors vs. number of folds")
plt.legend(['Training accuracy', 'Validation accuracy'])
plt.show()
```

```python
# This function is used to plot the confusion matrix and normalized confusion matrix
import itertools
from sklearn.metrics import confusion_matrix
from sklearn.metrics import confusion_matrix

def visualize_cm(cm):
    """
    Function visualizes a confusion matrix with and without normalization
    """
    plt.rc('legend', fontsize=10) 
    plt.rc('axes', labelsize=10) 
    plt.rc('xtick', labelsize=10) 
    plt.rc('ytick', labelsize=10) 

    fig, axes = plt.subplots(1, 2,figsize=(10,5))

    im1 = axes[0].imshow(cm, interpolation='nearest', cmap=plt.cm.Blues)
    fig.colorbar(im1, ax=axes[0])
    classes = ['Class 0','Class 1','Class 2','Class 3']
    tick_marks = np.arange(len(classes))
    axes[0].set_xticks(tick_marks)
    axes[0].set_xticklabels(classes,rotation=45)
    axes[0].set_yticks(tick_marks)
    axes[0].set_yticklabels(classes)

    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        axes[0].text(j, i, format(cm[i, j], 'd'),
                 horizontalalignment="center",
                 verticalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    axes[0].set_xlabel('predicted label $\hat{y}$')
    axes[0].set_ylabel('true label $y$')
    axes[0].set_title(r'Without normalization')
    
    cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
    im2 = axes[1].imshow(cm, interpolation='nearest', cmap=plt.cm.Blues)
    fig.colorbar(im2, ax=axes[1])
    
    axes[1].set_xticks(tick_marks)
    axes[1].set_xticklabels(classes,rotation=45)
    axes[1].set_yticks(tick_marks)
    axes[1].set_yticklabels(classes)

    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        axes[1].text(j, i, format(cm[i, j], '.2f'),
                 horizontalalignment="center",
                verticalalignment="center",
                color="white" if cm[i, j] > thresh else "black")

    axes[1].set_xlabel('predicted label $\hat{y}$')
    axes[1].set_ylabel('true label $y$')
    axes[1].set_title(r'Normalized')
    
    axes[0].set_ylim(-0.5,3.5) 
    axes[1].set_ylim(-0.5,3.5)
    
    plt.tight_layout()
    plt.show()
```

## 3.4  Multi-class logistic regression + Regularisation 

```python
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error
```

```python
train_fold5=fold5[0][0]
val_fold5=fold5[0][1]

df_train_K5=df_train[train_fold5]
df_val_K5=df_train[val_fold5]

y_train_K5=y_mul[train_fold5]
y_val_K5=y_mul[val_fold5]
```

```python
def regul_C_search(X_train, X_val, y_train, y_val, alpha_values):
    n_values = len(alpha_values)    # The number of candidate values for 'alpha'
    acc_train = np.zeros([n_values,1])    # Array for training errors
    acc_val = np.zeros([n_values,1])    # Array for validation errors

    for l in range(n_values):
        #penalty='l2'#2norm Ridge
        #penalty='l1'#1norm Lasso
        #One Vs Rest algorithm --> use 'liblinear' solver with it.
        #C:float, default=1.0
        logreg_reg = LogisticRegression(C=alpha_values[l], solver='liblinear', random_state=0, multi_class="ovr")
        logreg_reg.fit(X_train, y_train)
        y_train_pred =  logreg_reg.predict(X_train).reshape(-1, 1)
        acc_train[l] = metrics.accuracy_score(y_train_pred, y_train)
        
        y_val_pred = logreg_reg.predict(X_val).reshape(-1, 1)
        acc_val[l] =  metrics.accuracy_score(y_val_pred, y_val)

    best_alpha_idx = np.argmax(acc_val)
    logreg_reg = LogisticRegression(C=alpha_values[best_alpha_idx], solver='liblinear', 
                                    random_state=0, multi_class="ovr")
    logreg_reg =logreg_reg.fit(X_train, y_train)
    w_opt = logreg_reg.coef_
    
    return w_opt, acc_train, acc_val
```

```python
# Specify a list of values for C to be considered
C_val = np.array([0.01, 0.05, 0.2, 1, 3, 10, 1e2, 1e3])

w_opt, acc_trainK5, acc_valK5 = regul_C_search(df_train_K5, df_val_K5, y_train_K5, y_val_K5, C_val)
```

**Choose the value of C resulting in the highest (validation) accuracy.**

```python
# Plot the training and validation errors
plt.figure(figsize=(10,6)) 
plt.plot(C_val, acc_trainK5, marker='o', color='black', label='training accuracy')    # Plot training errors
plt.plot(C_val, acc_valK5, marker='o', color='red', label='validation accuracy')    # Plot validation errors
plt.xscale('log')    # Set x-axis to logarithmic scale
plt.xlabel(r'C')    # Set label of x-axis
plt.ylabel(r'Accuracy')    # Set label of y-axis
plt.title(r'Logistic REGR.+ Regularisation; Accuracies with respect to C', fontsize=16)    # Set title
plt.legend()    # Show legend
plt.show()
```

## 3.6 Difference between logistic regression and decision trees

The two classification methods logistic regression and decision trees both aim at learning a good predictor $h(\mathbf{x})$ which allows determining the label $y$ of the data point based on some features $\mathbf{x}$. These two classification methods differ in the form of predictor function $h(\mathbf{x})$ they are using. Logistic regression uses linear predictor functions $h(\mathbf{x})=\mathbf{w}^{T} \mathbf{x}$ (which are thresholded to get discrete label predictions $\hat{y}$). 

In contrast to linear functions used in logistic regression, decision trees use predictor functions that are obtained from flow charts (decision trees) consisting of various tests on the features $\mathbf{x}$. Using sufficiently large decision trees allows to represent highly non-linear functions $h(\mathbf{x})$. In particular, decision trees can perfectly separate data points (according to their labels) which cannot be separated by any straight line (which are the only possible decision boundaries for logistic regression).

##  Training on wide and deep model
- epoch = 50

```python
from tensorflow.keras.layers import Input, Dense, Concatenate
from tensorflow.keras.models import Model, Sequential

def NNModel(n_input, n_output):
    n_hidden1 = 120
    n_hidden2 = 60
    n_hidden3 = 30
    #n_output = 1
  
    input1 = Input(shape=n_input)
    hidden1 = Dense(units=n_hidden1, activation='elu')(input1)
    hidden2 = Dense(units=n_hidden2, activation='elu')(hidden1)
    hidden3 = Dense(units=n_hidden3, activation='elu')(hidden2)
    concat1 = Concatenate()([input1, hidden3])
    output1 = Dense(units=n_output, activation='linear')(concat1)

    model = Model(inputs=input1, outputs=output1, name='NN_model')
    
    model.compile(loss='mean_squared_error', optimizer='adam',
                  metrics=[tf.keras.metrics.MeanAbsolutePercentageError()]
                 #metrics=['accuracy']
                 )
    
    #model.compile(optimizer=tf.keras.optimizers.RMSprop(lr=2e-5),loss="categorical_crossentropy",
              #metrics=['accuracy'])

    return model
```

```python
# helper function for calculating MAPE
# Input: two numpy array with same shape (batch_size, )
# Output: MAPE in percentage(%)
def mape(y, y_predict):
    absSum = 0.0
    n = y.shape[0]
    for i in range(n):
        error = abs((y[i].item() - y_predict[i].item())/y[i].item())
        absSum += error
    return 100 * absSum/n
```

# 5. Conclusion

In The present study the chemical compounds of the hardened cement paste close to the fibre surface, such as \verb@'CSH_gel', 'High_Ca', 'CH', 'High_Si'@,  were predicted using the ML classification methods.
The data points were the discrete scans of the cement paste close to the fibre surface made by SEM-EDX method, delivering the chemical elements, such as Ca, Si and Al. 
Labels: reference ratios of the chemical elements in the CH and CSH gel

The classes were defined based on the reference ratios of the the Si/Ca in the CH and CSH gel received by the EDX individual micro-analysis of these compounds and reported in the previous similar studies \cite{eik.rsf.jact.2019}. The ratio of Si/Ca largely defines the presence of either CSH gel, providing the bond between the inclusion and the cement paste, or the CH having weak bonding properties. 

The original data was normalaised and centered around the zero.
The splitting of the data into training and validation sets was implemented using the 5-fold cross-validation. The number of the:

* training set samples was 2080;

* validation set samples 521;

* test set samples 434.

To select the best hypothesis space, the following models were tested:
\begin{itemize}
\item Multi-class Logistic Regression;
\item Regularisations with Lasso and Ridge regressions;
\item Decision Tree classifier;
\item Artificial Neural Network.
\end{itemize}

The best hypothesis selected was the Decision tree classifier, since it had the highest accuracy on the validation set, i.e. 97.31\%. On the test set, the accuracy of this model was 100\%.
The accuracy of the ANN model was also good, i.e. on the validation set it was 86.37\% and on the test set it was 85.02\%.

The weight values received in Logistic Regression demonstrated the relevance of each feature for the model. It was found that the most important feature for the Class 0 was  $x^1$ the distance from the fibre surface.
For the other Classes the most considerable feature was $x^2$ Si/Ca. The effect of the feature $x^3$ Al/Ca was low, however, it could also be considered while determining the classes. 

The clustering results with 4 and 6-means represented in Figure \ref{} demonstrate the correlation between the feautes $x^2$ Si/Ca and $x^3$ Al/Ca, which may support the assumption of assigning the classes by fulfilling the conditions of both ratios, i.e. Si/Ca and Al/Ca.
